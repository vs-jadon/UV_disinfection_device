```python
I1 = []

x = 6
### # Defining 'r' which is perpendicular distance between irradiance point and tube.
al = []
bl = []
X = []
Y = []
Z = []

R = []
# Caution: Above block must remain together with the below block, if not contents of X, Y, I lists will repeat as you rerun the block

for i in range (X1, X2):
    for j in range (Y1,Y2):   
        for k in range(Z1,Z2):
            X.append(i)
            Y.append(j)
            Z.append(k)
            
            if ((0<=i<3 and j>=0) or (0<=j<3 and i>=0) or (0<=i<3 and k>=0) or (0<=k<3 and j>=0) or (0<=j<3 and k>=0)
                or (0<=k<3 and j>=0) or (i>27) or (j>41) or (k>27) or (i>9 and j>7) or (i>18)):
                E = 0
            
           
            elif (i == 10 and j  > 5 ):
                E = 0
            elif (i == 11 and j > 4 ):
                E = 0
            elif (i == 12 and j > 3 ):
                E = 0
            elif (i == 13 and j > 2 ):
                E = 0
            elif (i > 13):
                E = 0
                    
                
            else: 
                r = math.sqrt(((x - i) ** 2) + (((z1 - k) ** 2)))
                R.append(r)
            
                ap = math.atan(((Ya - j)) / r) 
                bt = math.atan(((Yb - j)) / r)
            
#                 if (j<14):
#                     ap = ap
#                     bt = bt
               
#                 elif ((j>=14) and j<30):
#                     ap = -ap
#                     bt = +bt
                
#                 else:
#                     ap = -ap
#                     bt = -bt
                                                                                                  
                al.append(ap*(180/(22/7)))
                bl.append(bt*(180/(22/7)))
                
                
                
                I0 = p / (lg * (22 / 7) ** 2)
                Obt = (math.sin(2 * bt))
                Pap = math.sin(2 * ap)
           
                OP = Obt - Pap
           
                AA = 2 * (bt - ap) + OP
                E = (I0 / (4 * r)) * AA
            I1.append(E)
```
